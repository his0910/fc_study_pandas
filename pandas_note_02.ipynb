## 1. CSV 파일 읽어오기

**CSV 파일**이란?
* Comma Separated Value의 약어로써, 쉼표로 구분된 파일
* 엑셀을 로딩할 수도 있지만, 쉼표로 구분된 파일이 훨씬 가볍다
* [공공데이터 포털](https://www.data.go.kr/)에서도 CSV 포맷의 파일을 제공한다.
